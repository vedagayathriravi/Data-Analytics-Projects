2.start time:when does most of the accidents happen(is it during the day/night),which day,which month has more accidents,what is the trend over years(increasing/decreasing)

```python
df.columns #to check starttime column
```

```python
df.Start_Time #this is a string but we need to convert this as a date type
```

```python
df.Start_Time[0] #the below date is in yyyy-mm-dd
```

```python
pd.to_datetime(df.Start_Time)# converted as date type and saving it in a variable
```

```python
df.Start_Time=pd.to_datetime(df.Start_Time)
```

```python
df.Start_Time[0]
```

```python
sns.distplot(df.Start_Time.dt.hour,bins=24,kde=False,norm_hist=True)#as we can see in the below graph most of the accidents are happening at 7AM.
```

```python
sns.distplot(df.Start_Time.dt.dayofweek,bins=7,kde=False,norm_hist=True) #for week
```

```python
df.Start_Time.dt.dayofweek == 6 #is the distribution of accidents by hour on weekends as on weekdays?(here sunday)
```

```python
sundays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.distplot(sundays_start_time.dt.hour,bins=24,kde=False,norm_hist=True) #on sundays peak hours is at 10am
```

```python
mondays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 0]
sns.distplot(mondays_start_time.dt.hour,bins=24,kde=False,norm_hist=True)#on monday's the highest accident occured around 7am
```

```python
sns.distplot(df.Start_Time.dt.month,bins=12,kde=False,norm_hist=True) #too many accidents are happening in feb starting
```

```python
df.Start_Time.dt.year == 2016 #to check the accidents that are happened in 2016
```

```python
df_2016=df[df.Start_Time.dt.year == 2016]
sns.distplot(df_2016.Start_Time.dt.month,bins=12,kde=False,norm_hist=True)
```
